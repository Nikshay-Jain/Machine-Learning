## Data Set

```python
import numpy as np
import pandas as pd
from sympy import symbols, derive_by_array, Matrix, sympify
```

```python
data = pd.read_csv("housing.csv")
data = data[:100]
data.drop("ZN", axis=1, inplace=True)
data.drop("CHAS", axis=1, inplace=True)
data.drop("RAD", axis=1, inplace=True)
data.drop("B", axis=1, inplace=True)
data.drop("LSTAT", axis=1, inplace=True)
data
```

```python
y = data["MEDV"]
y = np.array(y)
data.drop("MEDV", axis = 1, inplace=True)
x = data.transpose()
x
```

```python
x = np.array(x)
```

```python
w1,w2,w3,w4,w5,w6,w7,w8,b = symbols('w1 w2 w3 w4 w5 w6 w7 w8 b', real = True)
w = np.array([w1,w2,w3,w4,w5,w6,w7,w8])
f = np.dot(w,x) + b - y
f
```

## Cost Function

```python
# Cost of entire table for given w[] and b
def cost(f):
    J = 1/(2*len(y))*np.sum(f**2)
    return J**0.5
```

## Gradient Descent

```python
def grad(w, b):
    alp = 0.0003
    w = w - alp*derive_by_array(cost(f),w)
    b = b - alp*derive_by_array(cost(f),b)
    return [w,b]
```

## Best fit

```python
t = 0
prev_cost = 0
num_w = np.zeros(8)
num_b = 1
gr_w, gr_b = grad(w, b)
gr_w = Matrix(gr_w)
while True:
    t+=1
    f = np.dot(num_w,x) + num_b - y
    
    num_w = gr_w.subs(w1,num_w[0]).subs(w2,num_w[1]).subs(w3,num_w[2]).subs(w4,num_w[3]).subs(w5,num_w[4]).subs(w6,num_w[5]).subs(w7,num_w[6]).subs(w8,num_w[7]).subs(b,num_b)
    num_b = gr_b.subs(w1,num_w[0]).subs(w2,num_w[1]).subs(w3,num_w[2]).subs(w4,num_w[3]).subs(w5,num_w[4]).subs(w6,num_w[5]).subs(w7,num_w[6]).subs(w8,num_w[7]).subs(b,num_b)
    
    num_w = np.reshape(num_w, (8,))
    if (abs(sympify(cost(f)).subs(w1,num_w[0]).subs(w2,num_w[1]).subs(w3,num_w[2]).subs(w4,num_w[3]).subs(w5,num_w[4]).subs(w6,num_w[5]).subs(w7,num_w[6]).subs(w8,num_w[7]).subs(b,num_b) - prev_cost) <= 0.001):
        break
    else:
        prev_cost = sympify(cost(f)).subs(w1,num_w[0]).subs(w2,num_w[1]).subs(w3,num_w[2]).subs(w4,num_w[3]).subs(w5,num_w[4]).subs(w6,num_w[5]).subs(w7,num_w[6]).subs(w8,num_w[7]).subs(b,num_b)
        print("RMS Cost:",prev_cost)
        print()
print("No of iterations:",t)
print("Coeff:",num_w)
print("Intercept:",num_b)
```
