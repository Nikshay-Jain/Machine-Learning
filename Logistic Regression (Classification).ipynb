## Sigmoid func

```python
def sigmoid(z):
    g = 1/(1 + np.exp(-z))
    return g
```

## Logistic Cost func

### J had 0*inf so unable to compute, so added smallest value of 10 to input of log that gave a finite output.

```python
def cost(x, y, w, b):
    J = 0
    for i in range(x.shape[0]):
        zi = np.dot(x[i],w) + b
        f_wb_i = sigmoid(zi)
        if y[i]==0:
            J+=(1-y[i])*np.log(1-f_wb_i)
        else:
            J+=y[i]*np.log(f_wb_i)
    J/=-13
    return J
```

## Gradient Descent

```python
def grad(w, b):
    alp = 0.0001
    dj_dw , dj_db = diff(x,y,w,b)
    w = w - alp*diff(cost(sigmoid(np.dot(x,w)+b)))
    b = b - alp*derive_by_array(cost(sigmoid(np.dot(x,w)+b)))
    return [w,b]
```

```python
def diff(X,y,w,b):
    m,n = X.shape
    dj_dw = np.zeros((n,))                           #(n,)
    dj_db = 0.

    for i in range(m):
        f_wb_i = sigmoid(np.dot(X[i],w) + b)          #(n,)(n,)=scalar
        err_i  = f_wb_i  - y[i]                       #scalar
        for j in range(n):
            dj_dw[j] = dj_dw[j] + err_i * X[i,j]      #scalar
        dj_db = dj_db + err_i
    dj_dw = dj_dw/m                                   #(n,)
    dj_db = dj_db/m                                   #scalar

    return dj_dw, dj_db
```
